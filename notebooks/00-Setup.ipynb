
# Workshop Setup

This notebook demonstrates the setup of **Credit Decision Scoring** workshop including the creation and management of catalogs, schemas, and managed volumes.

```python
CATALOG = 'workspace'
BRONZE_SCHEMA = 'bronze'
SILVER_SCHEMA = 'silver'
GOLD_SCHEMA = 'gold'

MANAGED_VOLUME_NAME = 'managed_vol_1'
```

```python
display(spark.sql(f"CREATE CATALOG IF NOT EXISTS {CATALOG};"))
display(spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{BRONZE_SCHEMA};"))
display(spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{SILVER_SCHEMA};"))
display(spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.{GOLD_SCHEMA};"))
```

```python
display(spark.sql(f"CREATE VOLUME IF NOT EXISTS {CATALOG}.default.{MANAGED_VOLUME_NAME};"))

# Upload the dataset to this volume in this path: /Volumes/{CATALOG}/default/{MANAGED_VOLUME_NAME}/credit_datasets
```
