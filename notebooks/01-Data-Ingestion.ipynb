```python
VOLUMEPATH = '/Volumes/apradana_demo_permata/default/managed_vol_1'
FILEPATH = VOLUMEPATH + '/credit_datasets'
```


# Read and Check Data

```python
# read applicants data

df_applicants = spark.read.csv(FILEPATH + '/1_data_nasabah.csv', header=True, inferSchema=True)
df_applicants.limit(5).display()
```

```python
# read loans data

df_loans = spark.read.csv(FILEPATH + '/2_data_pinjaman.csv', header=True, inferSchema=True)
df_loans.limit(5).display()
```

```python
# read payment data

df_repayments = spark.read.csv(FILEPATH + '/3_data_pembayaran.csv', header=True, inferSchema=True)
df_repayments.limit(5).display()
```


# Save data into table

```python
%sql
CREATE SCHEMA IF NOT EXISTS apradana_demo_permata.bronze;
```

```python
df_applicants.write.mode("overwrite").saveAsTable("apradana_demo_permata.bronze.raw_applicants")
df_loans.write.mode("overwrite").saveAsTable("apradana_demo_permata.bronze.raw_loans")
df_repayments.write.mode("overwrite").saveAsTable("apradana_demo_permata.bronze.raw_repayments")
```
